```python
import os
import pandas as pd
from collections import Counter

# Replace this with your directory containing the CSV files
directory_path = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\data_processed'

# List to keep track of each file's column count
file_column_counts = []

# Loop through all files in the directory
for file in os.listdir(directory_path):
    if file.endswith('.csv'):
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        num_columns = len(df.columns)
        file_column_counts.append((file, num_columns))

# Analyze the results to find the most common column count
column_count_frequencies = Counter(count for _, count in file_column_counts)
most_common_count = column_count_frequencies.most_common(1)[0][0]

# Print files with column counts different from the most common count and delete them
for file, count in file_column_counts:
    if count != most_common_count:
        unusual_file_path = os.path.join(directory_path, file)
        print(f"Deleting {file} as it has an unusual number of columns: {count}")
        os.remove(unusual_file_path)

print("Finished processing files.")
```

```python
import os
import pandas as pd
import numpy as np

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees).
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r

def calculate_distance_for_file(csv_file):
    # Load the CSV file
    df = pd.read_csv(csv_file)
    
    # Calculate the distances
    distances = [
        haversine(lon1, lat1, lon2, lat2) 
        for lat1, lon1, lat2, lon2 in zip(df['latitude'][:-1], df['longitude'][:-1], df['latitude'][1:], df['longitude'][1:])
    ]
    
    # Sum the distances
    return sum(distances)

def calculate_cumulative_distance(directory):
    total_distance = 0
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            file_path = os.path.join(directory, filename)
            total_distance += calculate_distance_for_file(file_path)
    return total_distance

# Example usage
directory = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\data_processed'  # Replace this with the path to your directory
cumulative_distance = calculate_cumulative_distance(directory)
print(f"Cumulative Distance Traversed: {cumulative_distance} km")
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Load your CSV file
csv_file_path = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\merged_file.csv'
df = pd.read_csv(csv_file_path)

# Assuming your columns are named 'latitude' and 'longitude'
latitude = 'latitude'
longitude = 'longitude'

# Create a map centered around the average location
map_center = [df[latitude].mean(), df[longitude].mean()]
m = folium.Map(location=map_center, zoom_start=10)

# Create a HeatMap layer and add it to the map
heat_data = [[row[latitude], row[longitude]] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

# Save the map to an HTML file
output_html = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\map.html'
m.save(output_html)

print(f"Heatmap saved to {output_html}")
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Load your CSV file
csv_file_path = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\merged_file.csv'
df = pd.read_csv(csv_file_path)

# Assuming your columns are named 'latitude' and 'longitude'
latitude = 'latitude'
longitude = 'longitude'

# Calculate the 10th and 90th percentiles for latitude and longitude to focus on the core 80%
lat_lower = df[latitude].quantile(0.30)
lat_upper = df[latitude].quantile(0.60)
long_lower = df[longitude].quantile(0.30)
long_upper = df[longitude].quantile(0.70)

# Filter the DataFrame to only include rows within these percentiles
core_df = df[(df[latitude] >= lat_lower) & (df[latitude] <= lat_upper) &
             (df[longitude] >= long_lower) & (df[longitude] <= long_upper)]

# Create a map centered around the average location of the core data
map_center = [core_df[latitude].mean(), core_df[longitude].mean()]
m = folium.Map(location=map_center, zoom_start=10)

# Create a HeatMap layer using core data and add it to the map
heat_data = [[row[latitude], row[longitude]] for index, row in core_df.iterrows()]
HeatMap(heat_data).add_to(m)

# Save the map to an HTML file
output_html = 'C:\\Users\\ss6365\\Desktop\\AR_GPS_Sensor_Data\\core_data_heatmap.html'
m.save(output_html)

print(f"Heatmap focusing on the core 80% of data saved to {output_html}")
```
